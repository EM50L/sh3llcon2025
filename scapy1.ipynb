## Paquete Ethernet (broadcast)

```python
# Crear el paquete Ethernet con carga útil
pkt = Ether(dst="ff:ff:ff:ff:ff:ff", type=0x2620) / Raw(load="Payload")
pkt.show()
hexdump(pkt)
```

```python
# Enviar el paquete y esperar respuesta durante 2 segundos
rsp = srp(pkt, iface="eth3", timeout=2, verbose=True)
# comprueba en el otro extremo que llega el paquete adb shell tcpdump -i eth0
```

## Paquete ARP

```python
# Crear un paquete ARP solicitando información del router
arp_req = ARP(pdst="172.32.0.70", op=1)
arp_req .show()
hexdump(arp_req)
```

```python
# Enviar el paquete y esperar respuesta durante 2 segundos
rsp = srp(arp_req, iface="eth3", timeout=5, verbose=True)
```

```python
from scapy.all import *

# Crear un paquete IP con un paquete ICMP dentro
packet = IP(dst="192.168.0.1")/ICMP(type=8, code=0)

# Enviar el paquete y recibir la primera respuesta
response = sr1(packet, verbose=0, timeout=2)

# Ver el resumen de la respuesta
if response:
    print("Respuesta recibida:",response.summary())
else:
    print("No se recibió respuesta")
```

```python
# consulta dns (equivale a dig +short SOA facebook.com @8.8.8.8)
from scapy.all import *

# Crear la consulta DNS
dnsQ = IP(dst="8.8.8.8")/UDP(dport=53)/DNS(rd=1, qd=DNSQR(qname="facebook.com", qtype="SOA"))

# Enviar la consulta y recibir la respuesta
resp = sr1(dnsQ, verbose=0)

# Mostrar la respuesta
if resp: resp.show()
else: print("No se recibió respuesta")
```
